```python
from datetime import datetime
from notebookutils import mssparkutils
import os

path = 'Files/pipeline_information/log_files/'
timestamp = datetime.now().strftime('%Y-%m-%d_%H-%M-%S')

try:
    raise Exception('Pipeline run has failed at {timestamp}!')
except Exception as e:
    file = 'Error_log_pipeline.txt'
    filepath = os.path.join(path, file)

    message = f'Error has occured at {timestamp} in pipeline run.\n'
    mssparkutils.fs.append(filepath, message, True)

    print(f'Error log entry has created at {timestamp}.')
```
